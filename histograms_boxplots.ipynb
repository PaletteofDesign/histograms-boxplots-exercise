```python
# Importing libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Importing data
f_name = '/home/spid3r404/Documents/data_science/data/catsvdogs.xlsx'
df = pd.read_excel(f_name)
df.head()
```

```python
# Graph showing percentage of dog owners
plt.style.use('seaborn')
plt.figure(figsize=(12,10))
df['Percentage of Dog Owners'].hist(bins=30, edgecolor='black', rwidth=0.7, color='green', label='Dogs')
plt.xlabel('Percentage of Dog Owners', labelpad=20, fontsize=14)
plt.ylabel('Count', labelpad=20, fontsize=14)
plt.title('Percentage of Dog Owners', pad=20, fontsize=16)
plt.show()
```

```python
# Graph showing percentage of cat owners
plt.style.use('seaborn')
plt.figure(figsize=(12,10))
df['Percentage of Cat Owners'].hist(bins=30, edgecolor='black', rwidth=0.7, label='Cats')
plt.title('Percentage of Cat Owners', fontsize=16, pad=20)
plt.xlabel('Percentage of Cat Owners', fontsize=14, labelpad=20)
plt.ylabel('Count', labelpad=20, fontsize=14)
plt.show()
```

```python
# Graph showing cat vs dog owners
plt.style.use('seaborn')
plt.figure(figsize=(12,10))
df['Percentage of Cat Owners'].hist(bins=30, edgecolor='black', rwidth=0.7, label= 'Cats')
df['Percentage of Dog Owners'].hist(bins=30, edgecolor='black', rwidth=0.7, label= 'Dogs')
plt.title('Cat vs Dog owners', fontsize=16, pad=20)
plt.xlabel('Percentage of Animal Owners', labelpad=20, fontsize=14)
plt.ylabel('Count', labelpad=20, fontsize=14)
plt.legend();
```

```python
# Similar to previous graph but classified as animal owners
plt.figure(figsize=(12,10))
sns.histplot(df['Percentage of Dog Owners'], color= 'green', label= 'Dogs')
sns.histplot(df['Percentage of Cat Owners'], color= 'blue', label= 'Cats')
plt.title('Percentage of Animal owners', fontsize=16, pad=20)
plt.xlabel('Percentage of Animal Owners', fontsize=14, labelpad=20)
plt.ylabel('Count', fontsize=14, labelpad=20)
plt.legend();
```

```python
# Box plots providing statistical representation of cat owners vs dog owners
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (12,10))
bp = axes.boxplot([df['Mean Number of Cats'],df['Mean Number of Dogs per household']], labels=['Cats', 'Dogs'], widths = .7, patch_artist=True, medianprops = dict(linestyle='-', linewidth=2, color='red'), boxprops = dict(linewidth=2, color='Black', facecolor = 'blue'))
boxplot1 = bp['boxes'][0]
boxplot1.set_facecolor('green')
plt.ylabel('Mean Number of Animals per State', fontsize = 16)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.show()
```

Evaluating the visual representations we can conclude that cats are more dominate as household pets over dogs.
